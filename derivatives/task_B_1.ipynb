```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt 
import matplotlib.ticker as ticker
from scipy.integrate import quad
from scipy.stats import norm
from tqdm import tqdm
```

### Dependence put option price on param 

```python
def d_plus(s, sigma, T, t, r, strike): 
    return ((np.log(s / strike)) + (T - t) * (r + (sigma ** 2) / 2)) / (sigma * np.sqrt(T - t))
```

```python
def d_minus(s, sigma, T, t, r, strike):
    return ((np.log(s / strike)) + (T - t) * (r - (sigma ** 2) / 2)) / (sigma * np.sqrt(T - t))
```

```python
def teta(s, sigma, T, t, r, strike):
    first = -1 *((s * sigma * np.exp(-(d_plus(s, sigma, T, t, r, strike) ** 2) / 2)) / (2 * np.sqrt(2 * math.pi) * (np.sqrt(T - t)))) 
    second = r * strike * np.exp(-r * (T - t)) * norm.cdf(-d_minus(s, sigma, T, t, r, strike))
    return first + second
```

#### Dependence on sigma


```python
T = 1
r = 0.1
strike = 1
t = 0
y_tetas = []
for sigma in np.linspace(0.01, 0.3, num=11):
    y_teta = []
    for s in np.linspace(0.0, 2 * strike, num=100):
        y_teta.append(teta(s, sigma, T, t, r, strike))
    y_tetas.append(y_teta)
```

```python
plt.rcParams['font.size'] = '15'
fig, ax = plt.subplots()

plt.plot(np.linspace(0.0, 2*strike, num=100), y_tetas[1], label='')
plt.plot(np.linspace(0.0, 2*strike, num=100), y_tetas[2], label='')
plt.plot(np.linspace(0.0, 2*strike, num=100), y_tetas[3], label='')
plt.plot(np.linspace(0.0, 2*strike, num=100), y_tetas[4], label='')
plt.plot(np.linspace(0.0, 2*strike, num=100), y_tetas[5], label='')
plt.plot(np.linspace(0.0, 2*strike, num=100), y_tetas[6], label='')
plt.plot(np.linspace(0.0, 2*strike, num=100), y_tetas[7], label='')

ax.grid(color = 'black',    #  цвет линий
            linewidth = 1,    #  толщина
            linestyle = '-')    #  начертание
plt.legend(loc = 'best', fancybox = True, shadow = True)

#plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(10))  # интервал написания значений оси OX
#plt.gca().yaxis.set_major_locator(ticker.MultipleLocator(10))  # интервал написания значений оси OY
plt.xticks(rotation=25)

fig.set_figwidth(10)
fig.set_figheight(5)


plt.title('teta, dependence on sigma')
plt.show()
```
