```python
T = 0.25
S = 100
r = 0.06
sigma = 0.3
u = 1.0395
d = 0.962
p = 0.4
q = 0.6
n = 15
STRIKE = 100
k = 7
```

price of the underlying asset

```python
def get_S(S_0, n):
    data_S = pd.DataFrame(np.zeros((n + 1, n + 1)))
    data_S.iloc[0, 0] = S_0
    for j in range(n + 1):
        for i in range(j, n + 1):
            data_S.iloc[j, i] = S * np.power(u, j) * np.power(d, i - j)
    return data_S
```

```python
data_S = get_S(100, 15)
```

```python
data_S
```

price of the eur option

```python
def get_payoff(S, STRIKE):
    return max(S - STRIKE, 0)
```

```python
[get_payoff(data_S.iloc[i, 15], STRIKE) for i in range(n)]
```

```python
def get_price_call_option(n, r, T, STRIKE):
    data_eur_option = pd.DataFrame(np.zeros(data_S.shape))
    data_eur_option[15] = [get_payoff(data_S.iloc[i, 15], STRIKE) for i in range(n + 1)]
    for j in range(n):
        for i in range(n - j, 0, -1):
            data_eur_option.iloc[i - 1, n - j - 1] = np.exp(-r * (T / n)) * (p * data_eur_option.iloc[i, n - j] + q * data_eur_option.iloc[i - 1, n - j])
    return data_eur_option
```

```python
data_eur_option = get_price_call_option(n, r, T, STRIKE)
```

```python
data_eur_option
```

price of the american option

```python
def get_amer_price_call_option(n, r, T):
    data_amer_option = pd.DataFrame(np.zeros(data_S.shape))
    data_amer_option[15] = data_eur_option[15]
    
    for j in range(n):
        for i in range(n - j, 0, -1):
            first_part = np.exp(-r * (T / n)) * (p * data_amer_option.iloc[i, n - j] + q * data_amer_option.iloc[i - 1, n - j])
            second_part = get_payoff(data_S.iloc[i - 1, n - j - 1])
            data_amer_option.iloc[i - 1, n - j - 1] = max(first_part, second_part)
    return data_amer_option
```

```python
data_amer_option = get_amer_price_call_option(n, r, T)
```

Futures

```python
def get_price_futures(k):
    data_futures = pd.DataFrame(np.zeros(( k + 1, k + 1)))
    data_futures[k] = data_S[k]
    for j in range(k):
        for i in range(k - j, 0, -1):
            data_futures.iloc[i - 1, k - j - 1] = (p * data_futures.iloc[i, k - j] + q * data_futures.iloc[i - 1, k - j])
    return data_futures
```

```python
data_futures = get_price_futures(k)
```

Option on futures

```python
def get_price_option_on_futures(k):
    data_option_on_futures = pd.DataFrame(np.zeros(data_futures.shape))
    data_option_on_futures[k] = list(map(get_payoff, data_futures[k]))

    for j in range(k):
        for i in range(k - j, 0, -1):
            first_part = np.exp(-r * (T / k)) * (p * data_option_on_futures.iloc[i, k - j] + q * data_option_on_futures.iloc[i - 1, k - j])
            second_part = get_payoff(data_futures.iloc[i - 1, k - j - 1])
            data_option_on_futures.iloc[i - 1, k - j - 1] = max(first_part, second_part)
    return data_option_on_futures
```

```python
data_option_on_futures = get_price_option_on_futures(k)
```
