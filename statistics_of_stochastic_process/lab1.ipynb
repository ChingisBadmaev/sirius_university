## Task 1


$$S_{n} = X_{1} + X_{2} + ... + X_{n}$$   
$$P({S_{[nt]} \over \sqrt(n)} < X) -> P(W(t)) < x$$    
$${S_{n} \over \sqrt(n)} -> N(0, 1)$$   


```python
def f(X, N, T):
    W =[]
    W += [0]
    Delta = T / N
    for i in range (1, N + 1):
        W += [W[i - 1] + np.random.normal(0, 1, 1) * np.sqrt(Delta)]
    return W
```

```python
N = 100
T = 1
X =[(i * T) / N for i in range(N + 1)]
W = f(X, N, T)
plt.figure(figsize=(10, 7))
plt.grid()
plt.plot(X ,W ,linewidth=2.0 )
plt.title('wiener_process')
plt.xlabel('t')
plt.ylabel('W')
plt.show()
```

### Task 2

$$ ||f||^{2} = \left(\int_0^T |f(t)|^{2}dt\right)^{1 \over 2}$$ 



$$W(t) = W(t, \omega) = \sum Z_{i}(\omega) \phi_{i}(t)$$ 

$$\phi_{i}(t) = {(2 \sqrt{2T}) \over ((2i + 1) \pi)} \sin({(2i + 1) \pi t \over 2T})$$

```python
def runif(n):
    L =[]
    L = np.sign(np.random.normal(0, 1, n))
    return L
```

```python
def function(x, n):
    S = np.cumsum(runif(n))
    if x * n > 0:
        return S[int(x * n ) -1]
    else :
        return 0
```

### Task 3

$$S(t) = x \exp ((r - {\sigma^{2} \over 2})t - \sigma W(t))$$   
$$S( t + \delta t) = S(t) \exp ((r - {\sigma^{2} \over 2})(t + \delta t -t) + \sigma (W(t + \delta t) - W(t))$$

```python
def phi (i, t, T):
    return(2 * np.sqrt(2 * T)) / ((2 * i + 1) * np.pi) * np.sin(((2 * i + 1) * np.pi * t ) / (2 * T))
```

```python
def sum_W (t , T, n, Z):
    Sum = 0
    for i in range(len(Z)) :
        Sum += Z [i] * phi(i ,t , T)
    return Sum
```

```python
plt.figure(figsize = (10, 7))
plt.grid()

T = 1
N = 100
t = [i / 100 for i in range(101)]

n = 10
Z = np.random.normal(0, 1, n)
W = []
for i in range(N + 1):
    W += [sum_W(t[i], T, N, Z)]
plt.plot (t , W , linewidth=2.0)

t = [i / 100 for i in range(101)]


n = 50
Z = np.random.normal(0, 1, n)
W = []
for i in range(N + 1):
    W += [sum_W(t[i], T, N, Z)]
plt.plot (t , W , linewidth=2.0)


t = [i / 100 for i in range(101)]

n = 100
Z = np.random.normal(0, 1, n)
W = []
for i in range(N + 1):
    W += [sum_W(t[i], T, N, Z)]
plt.plot (t , W , linewidth=2.0)
plt.title('Wiener_process')
plt.xlabel('t')
plt.ylabel('W')
plt.legend(['n = 10' , 'n = 50' , 'n = 100'])
```
